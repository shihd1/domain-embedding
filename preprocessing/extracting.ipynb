Extraction Pipeline

```python
from DocProcessing import DocProcessing

file_name = "../files/Forsthoffer's Vol 1 - Rotating Equipment.pdf"

processor = DocProcessing(filename=file_name)
extracted_text = processor.extractt_text()
filtered_text = processor.filter_lines(extracted_text)
```

Save Extracted Text

```python
name = "../extracted/" + file_name[9::]
with open(name, 'w+', encoding="utf-8") as file:
    file.write(filtered_text)
```
